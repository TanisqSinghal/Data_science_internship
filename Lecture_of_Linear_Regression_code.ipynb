## Outlier Analysis

```python
sns.boxplot(y="salary",data=df)
```

```python
percentile25=df['degree_p'].quantile(0.25)
```

```python
percentile75=df['degree_p'].quantile(0.75)
iqr=percentile75-percentile25
iqr
```

```python
uper_limit=percentile75+(1.5*iqr)
lower_limit=percentile25-(1.5*iqr)
```

```python
df=df[df['degree_p']<uper_limit]
df=df[df['degree_p']>lower_limit]
```

```python
col=['ssc_p','hsc_p','degree_p','etest_p','mba_p']
```

```python
for c in col:
    percentile25=df[c].quantile(0.25)
    percentile75=df[c].quantile(0.75)
    iqr=percentile75-percentile25
    uper_limit=percentile75+(1.5*iqr)
    lower_limit=percentile25-(1.5*iqr)
    df=df[df[c]<uper_limit]
    df=df[df[c]>lower_limit]
    plt.figure()
    sns.boxplot(y=c,data=df)
```

## Linear Regression

```python
X=df.iloc[:,0:-2].values
```

```python
X
```

```python
Y=df.iloc[:,-1].values
```

```python
Y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0,2,4,5,7,8,10])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2, random_state=4)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Feature Scaling

```python
## Feature scaling is used if one column values are more in width then others column and it will dominate in the model /
## just like salary is more in width then hss_b here so we have to bring them in same
```

## Two types of feature Scaling

## 1. Normalization and to convert it has formula of  :- -------                                x-xmin / Xmax-Xmin
## 2. Standardization and to convert it has formula of  :-------                    x-mean of column / standard division of column

```python
from sklearn.preprocessing import MinMaxScaler
mm =MinMaxScaler()
```

```python
X_train=mm.fit_transform(X_train)
```

```python
X_test=mm.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train, Y_train)
```

```python
Y_pred=reg.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
reg.score(Y_test, Y_pred)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(Y_test,Y_pred)
```

```python
import numpy as np
np.sqrt(metrics.mean_squared_error(Y_test,Y_pred))
```

```python
metrics.r2_score(Y_test,Y_pred)
```

```python
import pandas as pd
import joblib

joblib.dump(reg, 'linear_regression_model.pkl')
# Load new data
new_data = np.array([ 0.76323278,0.76323278, -1.05682691,  1.05682691, -0.82487112,  0.82487112,
       -0.21182964,  0.91184465, -0.83541407,  0.67480156, -0.19549056,
       -0.61754023, -1.34549055,  1.34549055,  0.88949918, -0.88949918,
        0.7528957 ,  0.97372676,  0.12192832,  0.10159344,  0.2852968

    # Add all your features here
                    ]).reshape(1, -1)
# Load the saved model
model = joblib.load('linear_regression_model.pkl')

# Make predictions
predictions = model.predict(new_data)

# Print the predictions
print(predictions)

```

```python
Y_train[1]
```

```python


# Show coefficients and intercept
intercept =  model.intercept_
coefficients  = model.coef_

print(intercept)
print(coefficients)
```

```python
features = [f"x{i}" for i in range(len(coefficients))]

equation = f"y = {intercept:.2f} "
for i, coef in enumerate(coefficients):
    sign = "+" if coef >= 0 else "-"
    equation += f"{sign} {abs(coef):.2f}*{features[i]} "

print("Regression Equation:")
print(equation)
```

```python


feature_names = [f"x{i}" for i in range(len(coefficients))]

plt.figure(figsize=(12, 6))
plt.bar(feature_names, coefficients)
plt.axhline(0, color='black', linewidth=0.8)
plt.xticks(rotation=45)
plt.title("Feature Coefficients in Linear Regression Model")
plt.xlabel("Features")
plt.ylabel("Coefficient Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(X)

# Example evaluation
print("Mean Squared Error:", mean_squared_error(Y_test, Y_pred))
print("R Score:", r2_score(Y_test, Y_pred))

# Optional: Scatter plot of predictions
plt.scatter(Y_test, Y_pred, color='purple')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--')  # 45-degree line
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.grid(True)
plt.show()
```
