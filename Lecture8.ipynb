#INTRO TO MATPLOTLIB

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
years= range(2000,2012)
apples = [0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937,0.9375,0.9372,0.939]
oranges=[0.992,0.941,0.930,0.923,0.918,0.908,0.907,0.904,0.901,0.898,0.9,0.896]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in haryana")
plt.legend(["Apples", "Oranges"])
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in haryana", loc="left") #position left of the title
plt.legend(["Apples", "Oranges"])
```

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in haryana")
plt.legend(["Apples", "Oranges"])
```

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=6, mew=2, mec='navy')
plt.plot(years, oranges, marker='x', c='r', ls='--', lw=1, ms=6, mew=2, alpha=0.8)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in haryana")
plt.legend(["Apples", "Oranges"])
```

```python
plt.plot(years, apples, marker='s', c='y', ls='-', lw=1, ms=5, mew=2, mec='navy')
plt.plot(years, oranges, marker='x', c='purple', ls='--', lw=1, ms=5, mew=2, alpha=0.8)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in haryana")
plt.legend(["Apples", "Oranges"])
```

```python
plt.plot(years, apples,"s--b")
plt.plot(years, oranges, "x--y")
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in haryana")
plt.legend(["Apples", "Oranges"])
```

```python
plt.figure(figsize=(2,2))

plt.plot(years, oranges, 'o-r')
plt.title("yield of oranges (tons per hectare)")
```

#HEATMAP
1. it provides info of data wrt x and y axis like how many passengers in every year in every month

```python
df=sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")
# print(df)
plt.title("No. of passengers (1000s)")
sns.heatmap(df, fmt='d',linecolor='black',linewidth=2, annot=True, cmap='Blues')
```

#Point plot

```python
df = sns.load_dataset("flights")

plt.title("Average passengers as per year")

sns.pointplot(x="year", data=df, y="passengers", hue="month")
```

#pie chart

```python
sns.set_theme(style="darkgrid")

colors = sns.color_palette('Set3', 12)
plt.figure(figsize=(8,8))
plt.pie(df.groupby("year")["passengers"].mean().astype("int"),
        labels=df.year.unique(),
        colors = colors, autopct="%1.1f%%",
        startangle=220, wedgeprops={"width" : 0.6})
plt.title("tips distribution year wise")
plt.show()
```

```python
sns.catplot(x="year", y="passengers", data=df, hue="month")
plt.figure(figsize=(12,6))
plt.show()
```

```python
sns.categorical.boxenplot(df.passengers, fill=True)
plt.show()
```

```python
sns.barplot(x="month",y="passengers", data=df,hue="month", palette="Set3", errorbar=None,estimator="sum")
```

```python
sns.displot(x="month", y="passengers", hue="year", data=df, palette="Set3", height=7)
plt.title("Month wise total passengers")
```

```python
sns.countplot(x="month", data=df, hue="month")
```

```python
sns.violinplot(x="year",y="passengers", data=df, palette="Set2", legend=True)
plt.show()
```
